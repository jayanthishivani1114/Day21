```python
# Day21 – Intro to Machine Learning

## Supervised Learning
Supervised Learning is a type of Machine Learning where the dataset contains input-output pairs (labels). 
The model learns by comparing its predictions to the correct answers and adjusting itself to minimize errors.
    It is commonly used for classification (predicting categories) and regression (predicting numerical values).

## Unsupervised Learning
Unsupervised Learning works with data that has no labels or predefined answers.
    The model tries to discover hidden patterns, structures, or clusters in the data. 
    It is commonly used for clustering and dimensionality reduction.

## Reinforcement Learning
Reinforcement Learning is a learning method where an agent interacts with an environment and learns by receiving rewards or
penalties based on its actions. The objective is to learn a policy that maximizes long-term rewards.

---

## Comparison Table

| Learning Type | Definition | Real-World Example 1 | Real-World Example 2 |
|---------------|------------|----------------------|----------------------|
| Supervised Learning | Learns from labeled data (inputs + correct outputs). | Spam Detection (Email → Spam/Not Spam) 
| House Price Prediction (Features → Price) |
| Unsupervised Learning | Finds hidden patterns in unlabeled data. | Customer Segmentation | Fraud/Anomaly Detection |
| Reinforcement Learning | Learns via rewards and penalties through interaction. | Game-playing AI (AlphaGo learning Go) |
Self-driving car decision systems |

---

## Why Choosing the Right Type Matters
Choosing the wrong learning framework can cause a project to fail before it even begins. 
    If you try supervised learning without labeled data, your model has nothing to learn from.
If you use unsupervised learning when labels exist, you waste valuable information.
    Reinforcement learning requires interaction and rewards—without them, it cannot function properly. 
    Correctly framing the problem is critical to success in Machine Learning.
```
